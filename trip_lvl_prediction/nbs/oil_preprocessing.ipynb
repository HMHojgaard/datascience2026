### Load in data

```python
oil_df = pd.read_csv(data_path / "DCOILBRENTEU.csv")
```

### Clean data

```python
# rename columns
oil_df = oil_df.rename(columns={
    "observation_date": "date",
    "DCOILBRENTEU": "oil_price"
})

# convert date column to datetime
oil_df["date"] = pd.to_datetime(oil_df["date"])

oil_df.head()
```

```python
# check for missing values
oil_df["oil_price"] = oil_df["oil_price"].ffill()
```

```python
oil_df.isna().sum()
```

```python
# create complete daily date range
full_dates = pd.DataFrame({
    "date": pd.date_range(
        start=oil_df["date"].min(),
        end=oil_df["date"].max(),
        freq="D"
    )
})

# merge oil data onto complete date range
oil_df_daily = full_dates.merge(
    oil_df,
    on="date",
    how="left"
)

# forward-fill missing oil prices
oil_df_daily["oil_price"] = oil_df_daily["oil_price"].ffill()

# check missing values
oil_df_daily.isna().sum()
```
